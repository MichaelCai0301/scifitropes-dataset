```python
# Read the cleaned CSVs into DataFrames
scififilmdb_cleaned = pd.read_csv("scififilmdb_cleaned.csv")
scifianimedb_cleaned = pd.read_csv("scifianimedb_cleaned.csv")
scifivideogamesdb_cleaned = pd.read_csv("scifivideogamesdb_cleaned.csv")

# Read the additional CSVs for extra columns
videogames_with_popularity_score = pd.read_csv("videogames_with_popularity_score.csv")
films_with_box_office = pd.read_csv("films_with_box_office.csv")
anime_with_popularity_score = pd.read_csv("anime_with_popularity_score.csv")

# Append 'influence_score' from videogames_with_popularity_score to scifivideogamesdb_cleaned
if 'influence_score' in videogames_with_popularity_score.columns:
    scifivideogamesdb_cleaned['influence_score'] = videogames_with_popularity_score['influence_score']

# Append 'BoxOfficeUSD' from films_with_box_office to scififilmdb_cleaned
if 'BoxOfficeUSD' in films_with_box_office.columns:
    scififilmdb_cleaned['BoxOfficeUSD'] = films_with_box_office['BoxOfficeUSD']

# Append 'popularity_score' from anime_with_popularity_score to scifianimedb_cleaned
if 'popularity_score' in anime_with_popularity_score.columns:
    scifianimedb_cleaned['popularity_score'] = anime_with_popularity_score['popularity_score']
```

```python
def plot_distributions_for_year(year):
    import matplotlib.pyplot as plt

    # Filter videogames for the given year
    vg_df = scifivideogamesdb_cleaned
    if 'Release Year' in vg_df.columns:
        vg_year = vg_df[vg_df['Release Year'] == year]
    elif 'release_year' in vg_df.columns:
        vg_year = vg_df[vg_df['release_year'] == year]
    else:
        vg_year = vg_df[vg_df.get('year', pd.Series([None]*len(vg_df))) == year]
    vg_scores = vg_year['influence_score'].dropna()

    # Filter films for the given year
    film_df = scififilmdb_cleaned
    if 'Release Year' in film_df.columns:
        film_year = film_df[film_df['Release Year'] == year]
    elif 'release_year' in film_df.columns:
        film_year = film_df[film_df['release_year'] == year]
    else:
        film_year = film_df[film_df.get('year', pd.Series([None]*len(film_df))) == year]
    film_box = film_year['BoxOfficeUSD'].dropna()

    # Filter anime for the given year
    anime_df = scifianimedb_cleaned
    if 'Release Year' in anime_df.columns:
        anime_year = anime_df[anime_df['Release Year'] == year]
    elif 'release_year' in anime_df.columns:
        anime_year = anime_df[anime_df['release_year'] == year]
    else:
        anime_year = anime_df[anime_df.get('year', pd.Series([None]*len(anime_df))) == year]
    anime_pop = anime_year['popularity_score'].dropna()

    fig, axs = plt.subplots(1, 3, figsize=(18, 5))

    # Videogames
    axs[0].hist(vg_scores, bins=100, color='skyblue', edgecolor='black')
    axs[0].set_title(f'Videogame Influence Score Distribution ({year})')
    axs[0].set_xlabel('Influence Score')
    axs[0].set_ylabel('Count')

    # Films
    axs[1].hist(film_box, bins=100, color='salmon', edgecolor='black')
    axs[1].set_title(f'Film Box Office Distribution ({year})')
    axs[1].set_xlabel('Box Office (USD)')
    axs[1].set_ylabel('Count')

    # Anime
    axs[2].hist(anime_pop, bins=100, color='lightgreen', edgecolor='black')
    axs[2].set_title(f'Anime Popularity Score Distribution ({year})')
    axs[2].set_xlabel('Popularity Score')
    axs[2].set_ylabel('Count')

    plt.tight_layout()
    plt.show()

def plot_distributions_for_10years(start_year):
    import matplotlib.pyplot as plt
    years = list(range(start_year, start_year + 10))
    
    vg_df = scifivideogamesdb_cleaned
    film_df = scififilmdb_cleaned
    anime_df = scifianimedb_cleaned

    # Helper to get values for a range of years
    def get_col_for_years(df, col, year_col_names):
        for ycol in year_col_names:
            if ycol in df.columns:
                mask = df[ycol].isin(years)
                return df.loc[mask, col].dropna()
        # fallback: try .get
        mask = df.get('year', pd.Series([None]*len(df))).isin(years)
        return df.loc[mask, col].dropna()

    # Videogames
    vg_scores = get_col_for_years(vg_df, 'influence_score_norm', ['Release Year', 'release_year'])

    # Films
    film_box = get_col_for_years(film_df, 'BoxOfficeUSD_norm', ['Release Year', 'release_year'])

    # Anime
    anime_pop = get_col_for_years(anime_df, 'popularity_score_norm', ['Release Year', 'release_year'])

    fig, axs = plt.subplots(1, 3, figsize=(18, 5))

    # Videogames
    axs[0].hist(vg_scores, bins=100, color='skyblue', edgecolor='black')
    axs[0].set_title(f'Videogame Influence Score Distribution ({years[0]}–{years[-1]})')
    axs[0].set_xlabel('Influence Score')
    axs[0].set_ylabel('Count')

    # Films
    axs[1].hist(film_box, bins=100, color='salmon', edgecolor='black')
    axs[1].set_title(f'Film Box Office Distribution ({years[0]}–{years[-1]})')
    axs[1].set_xlabel('Box Office (USD)')
    axs[1].set_ylabel('Count')

    # Anime
    axs[2].hist(anime_pop, bins=100, color='lightgreen', edgecolor='black')
    axs[2].set_title(f'Anime Popularity Score Distribution ({years[0]}–{years[-1]})')
    axs[2].set_xlabel('Popularity Score')
    axs[2].set_ylabel('Count')

    plt.tight_layout()
    plt.show()
```

```python
vg_df = scifivideogamesdb_cleaned
film_df = scififilmdb_cleaned
anime_df = scifianimedb_cleaned

# Normalize influence_score in vg_df
if 'influence_score' in vg_df.columns:
    min_inf = vg_df['influence_score'].min()
    max_inf = vg_df['influence_score'].max()
    vg_df['influence_score_norm'] = (vg_df['influence_score'] - min_inf) / (max_inf - min_inf)

# Normalize BoxOfficeUSD in film_df
if 'BoxOfficeUSD' in film_df.columns:
    min_box = film_df['BoxOfficeUSD'].min()
    max_box = film_df['BoxOfficeUSD'].max()
    film_df['BoxOfficeUSD_norm'] = (film_df['BoxOfficeUSD'] - min_box) / (max_box - min_box)

# Normalize popularity_score in anime_df
if 'popularity_score' in anime_df.columns:
    min_pop = anime_df['popularity_score'].min()
    max_pop = anime_df['popularity_score'].max()
    anime_df['popularity_score_norm'] = (anime_df['popularity_score'] - min_pop) / (max_pop - min_pop)
```

```python


from collections import defaultdict
import ast

film_tropes = defaultdict(dict)

# Ensure required columns exist
required_cols = {'Release Year', 'Tropes', 'BoxOfficeUSD_norm'}
if required_cols.issubset(film_df.columns):
    # Drop rows with missing values in required columns
    film_df_years = film_df.dropna(subset=list(required_cols)).copy()
    # Convert 'Release Year' to int safely
    film_df_years['Release Year'] = film_df_years['Release Year'].astype(int)
    # Group by year
    for year, group in film_df_years.groupby('Release Year'):
        trope_boxoffice = defaultdict(float)
        for _, row in group.iterrows():
            # Parse the Tropes column (should be a stringified dict)
            tropes_str = row['Tropes']
            try:
                tropes_dict = ast.literal_eval(tropes_str) if isinstance(tropes_str, str) else {}
                if isinstance(tropes_dict, dict):
                    for trope in tropes_dict:
                        # Add BoxOfficeUSD for each trope
                        box_office = row['BoxOfficeUSD_norm']
                        if isinstance(box_office, (int, float)):
                            trope_boxoffice[trope] += box_office
            except Exception:
                continue
        film_tropes[year] = dict(trope_boxoffice)

```

```python

# For videogames: aggregate normalized influence score by trope per year
vg_tropes = defaultdict(dict)

required_cols_vg = {'Release Year', 'Tropes', 'influence_score_norm'}
if required_cols_vg.issubset(vg_df.columns):
    vg_df_years = vg_df.dropna(subset=list(required_cols_vg)).copy()
    vg_df_years['Release Year'] = vg_df_years['Release Year'].astype(int)
    for year, group in vg_df_years.groupby('Release Year'):
        trope_influence = defaultdict(float)
        for _, row in group.iterrows():
            tropes_str = row['Tropes']
            try:
                tropes_dict = ast.literal_eval(tropes_str) if isinstance(tropes_str, str) else {}
                if isinstance(tropes_dict, dict):
                    for trope in tropes_dict:
                        influence = row['influence_score_norm']
                        if isinstance(influence, (int, float)):
                            trope_influence[trope] += influence
            except Exception:
                continue
        vg_tropes[year] = dict(trope_influence)

# For anime: aggregate normalized popularity score by trope per year
anime_tropes = defaultdict(dict)

required_cols_anime = {'Release Year', 'Tropes', 'popularity_score_norm'}
if required_cols_anime.issubset(anime_df.columns):
    anime_df_years = anime_df.dropna(subset=list(required_cols_anime)).copy()
    anime_df_years['Release Year'] = anime_df_years['Release Year'].astype(int)
    for year, group in anime_df_years.groupby('Release Year'):
        trope_popularity = defaultdict(float)
        for _, row in group.iterrows():
            tropes_str = row['Tropes']
            try:
                tropes_dict = ast.literal_eval(tropes_str) if isinstance(tropes_str, str) else {}
                if isinstance(tropes_dict, dict):
                    for trope in tropes_dict:
                        popularity = row['popularity_score_norm']
                        if isinstance(popularity, (int, float)):
                            trope_popularity[trope] += popularity
            except Exception:
                continue
        anime_tropes[year] = dict(trope_popularity)

```

```python



def dict_to_sorted_df(trope_dict, score_col_name):
    # trope_dict: {year: {trope: score, ...}, ...}
    # Returns: {year: DataFrame sorted by score_col_name descending}
    result = {}
    for year, tropes in trope_dict.items():
        df = pd.DataFrame(list(tropes.items()), columns=['Trope', score_col_name])
        # Convert score column to numeric, coerce errors to NaN, then drop NaNs
        df[score_col_name] = pd.to_numeric(df[score_col_name], errors='coerce')
        df = df.dropna(subset=[score_col_name])
        df = df.sort_values(by=score_col_name, ascending=False).reset_index(drop=True)
        result[year] = df
    return result

# For film
film_tropes_df = dict_to_sorted_df(film_tropes, 'score')

# For anime
anime_tropes_df = dict_to_sorted_df(anime_tropes, 'score')

# For video games
vg_tropes_df = dict_to_sorted_df(vg_tropes, 'score')

```

```python
def topXTropes(df, year, x):
    # get the top x tropes of given year, provided the df type
    return df[year].nlargest(x, 'score')
```
